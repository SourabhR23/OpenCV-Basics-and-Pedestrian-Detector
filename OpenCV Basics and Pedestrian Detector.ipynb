### Basics of Computer Vision

#### Read the image

```python
# Importing the OpenCV library
import cv2
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure, imshow

# Reading the image using imread() function
image = cv2.imread('image.jpg')

# Displaying the image
plt.axis("off")
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
# Extracting the height and width of the image
h, w = image.shape[:2]
print(f"Height = {h}, Width = {w}")
```

#### Resizing the Image

```python
resize = cv2.resize(image, (500, 720))
plt.axis("off")
plt.imshow(cv2.cvtColor(resize, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
# Calculating the ratio
ratio = 720 / w

# Creating a tuple containing width and height
dim = (500, int(h * ratio))

# Resizing the image
resize_aspect = cv2.resize(image, dim)

# Calculating the center of the image
center = (w // 2, h // 2)
```

```python
center
```

#### Rotating the Image

```python
# Generating a rotation matrix
matrix = cv2.getRotationMatrix2D(center, -45, 1.0)

# Performing the affine transformation
rotated = cv2.warpAffine(image, matrix, (w, h))

# Displaying the image
plt.axis("off")
plt.imshow(cv2.cvtColor(rotated, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
# Generating a rotation matrix
matrix = cv2.getRotationMatrix2D(center, 65, 0.8)

# Performing the affine transformation
rotated = cv2.warpAffine(image, matrix, (w, h))

# Displaying the image
plt.axis("off")
plt.imshow(cv2.cvtColor(rotated, cv2.COLOR_BGR2RGB))
plt.show()
```

#### Drawing the rectangle

```python
# Copying the original image
op = image.copy()

# Using the rectangle() function to create a rectangle(bounding box)
rect = cv2.rectangle(op, (87, 8), (552, 410), (255, 0, 0), 2)

# Displaying the image
plt.axis("off")
plt.imshow(cv2.cvtColor(op, cv2.COLOR_BGR2RGB))
plt.show()
```

### Pedestrian Detector

```python
import cv2
import imutils

# Initializing the HOG person detector
hog = cv2.HOGDescriptor()
hog.setSVMDetector(cv2.HOGDescriptor_getDefaultPeopleDetector())

# Reading the Image
image = cv2.imread('walk.jpg')

# Resizing the Image
image = imutils.resize(image, width = min(400, image.shape[1]))

# Detecting all the regions in the image that has a pedestrians inside it
(regions, _) = hog.detectMultiScale(image, winStride = (2, 2), padding = (4, 4), scale = 1.05)

# Drawing the regions in the Image
for (x, y, w, h) in regions:
    cv2.rectangle(image, (x, y), (x + w, y + h), (0, 0, 255), 2)

# Displaying the image
plt.axis("off")
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
```
